# Exploratory Data Analysis on IPL Dataset

### Importing the necessary libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Reading and Displaying the data

```python
data = pd.read_csv("C:\\Users\\ramana777\\Downloads\\matches.csv")
print(data)
```

### Basic Data Exploration

```python
print("Dataset First Five Rows:")
print(data.head())
```

```python
print("Dataset Info:")
print(data.info())
```

```python
print("Dataset Statistics:")
print(data.describe())
```

```python
print(data.describe(include=['object']))
```

```python
numerical_columns = data.select_dtypes(include=["int64","float64"]).columns
for col in numerical_columns:
    print(f"\nStatistics of {col}:")
    print(data[col].describe())
```

```python
categorical_columns = data.select_dtypes(include=["object"]).columns
for col in categorical_columns:
    print(f"\nStatistics of {col}:")
    print(data[col].describe())
```

### Data Visualizations

```python
for col in numerical_columns:
    plt.figure(figsize=(10,6))
    sns.histplot(data[col],kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(data.corr(numeric_only=True),annot=True,cmap="coolwarm",fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.countplot(x='toss_decision',hue='winner',data=data)
plt.title("Impact of Toss Decision on Match Outcome")
plt.show()
```

```python
win_counts = data["winner"].value_counts()
plt.figure(figsize=(10,6))
win_counts.plot(kind='bar',color='skyblue')
plt.title("Number of Matches Won by Each Team")
plt.ylabel("Wins")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='season', data=data, palette='viridis')
plt.title('Number of Matches per Season')
plt.xticks(rotation=90)
plt.show()
```

```python
sns.pairplot(data[numerical_columns].dropna())
plt.show()
```

```python
win_rate = data["winner"].value_counts(normalize=True) * 100
plt.figure(figsize=(10,6))
win_rate.plot(kind='bar',color='purple')
plt.title("Win Percentage by Team")
plt.ylabel("Win Percentage")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12,12))
sns.countplot(y=data["venue"],order=data["venue"].value_counts().index,palette="coolwarm")
plt.title("Perfomance at Different Venues")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(y=data["player_of_match"],order=data["player_of_match"].value_counts().iloc[:10].index,palette="magma")
plt.title('Top 10 Players with Most "Player of the Match" Awards')
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x=data['result_margin'].dropna(), color='green')
plt.title('Boxplot of Result Margin')
plt.show()
```

```python
toss_vs_win = data.groupby('toss_decision')['winner'].value_counts(normalize=True).unstack() * 100
toss_vs_win.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Toss Decision Impact on Winning Probability')
plt.ylabel('Winning Percentage')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='toss_decision', hue='winner', data=data, palette='Set1')
plt.title('Wins by Batting First vs Chasing')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='super_over', hue='winner', data=data)
plt.title('Impact of Super Overs on Winning')
plt.show()
```
